```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
# X_train[0]
# all the values are between the 0 and 255 we have to take them into a range so the convergence of the weights should be faster more enough.

X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
# now for input layer i have the inputs in the format like 28 * 28 i have to take it to the shape like 1,784 row by row so i can feed these inputs into my model
model = Sequential()
# Flatten from keras is used to change the high dimensinol data to 1D.
model.add(Flatten(input_shape=(28,28)))
# will going to create the neural network
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=30, validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history[
  'val_loss'
])
```
